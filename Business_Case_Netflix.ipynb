```python
continuous_vars = ['duration','rating','release_year']
```

```python
for var in continuous_vars:
  # distplot
  sns.displot(data[var], kde=True)
  plt.title(f'Distribution of {var}')
  plt.show()
```

```python
# countplot
sns.countplot(data[var])
plt.title(f'Distribution of {var}')
plt.show()
```
